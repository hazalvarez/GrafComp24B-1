# Triangulación 
# Número de vertices n=16
## por lo tanto el numero de tringulos es n-2 = 16-2 = 14
### El poligono se marca en color negro, mientra que las diagonales se marcan de color rojo


```python
#1
x1 = 0
x2 = 30
y1 = 0
y2 = 30
plt.axis([x1, x2, y1, y2])
#2
plt.grid(True)
#3
plt.plot([15.5,13,5,0,1.5,5,7,4.7,5.3,14.5,13.5,8,12,17.5,17,25,15.5],[6,2,0,7,15,11,13,15.5,19.7,18,15,10.5,7,11.7,15.5,8,6], 'k-')

##Se realiza la Triangulación
plt.plot([0, 5], [7, 11], 'r-')     
plt.plot([5, 5], [11, 0], 'r-') 
plt.plot([5.3, 13.5], [19.7, 15], 'r-')
plt.plot([13.5, 4.7], [15, 15.5], 'r-') 
plt.plot([25, 17.5], [8, 11.7], 'r-') 
plt.plot([17.5, 15.5], [11.7, 6], 'r-')  
plt.plot([15.5, 12], [6, 7], 'r-')     
plt.plot([12, 13], [7, 2], 'r-')       
plt.plot([12, 5], [7, 0], 'r-')        
plt.plot([5, 8], [0, 10.5], 'r-')      
plt.plot([8, 5], [10.5, 11], 'r-')      
plt.plot([8,7], [10.5,13], 'r-')      
plt.plot([7,13.5], [13,15], 'r-')  

plt.show()

```
